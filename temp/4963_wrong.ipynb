```python
def parse_input_string(string):
    return [int(number) for number in string.split(' ') if int(number) == 1]


txt_file = '4963_txt.txt'
# def main():
with open(txt_file, "r") as f:
    
    # infinite Loop
    while True:
        w, h = map(int, f.readline().split())
        if w == 0 and h == 0:
            break
        
        print("\nW and H : ",w, h)

        results = []
        result = 0
        for i in range(h):
            # Input string이 주어짐
            input_string = f.readline()
            print(input_string)

            # Input String를 Int list로 parsing함
            int_list = parse_input_string(input_string)

            # 모든 섬인 경우를 더한다.
            result += sum(int_list)
            print(result)
            print("=====")
            
            # 결과를 list에 담음
            results.append(result)
            

#             print("\n".join(results))
```
